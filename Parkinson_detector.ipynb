```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Reading the files

```python
# Accessing the data
data=pd.read_csv('parkinsons data.csv')
```

```python
# dropping columns with null values 
data.dropna(axis=1, inplace=True)
```

## Using graphs for better understanding

```python
sns.heatmap(data.drop(columns=['status', 'name']).corr(), cmap='coolwarm', annot=True)
```

## Train test split

```python
y=data['status']
X=data[['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR','RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE']]
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(random_state=1)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=104)
```

```python
model.fit(X_train,y_train)
```

## Predicting the test data

```python
prediction=model.predict(X_test)
```

```python
print(model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
valpredict=model.predict(X_test)
```

```python
print(mean_absolute_error(y_test,valpredict))
```

```python
from sklearn.metrics import classification_report
```

## Comparing the result with actual answers

```python
print(classification_report(y_test,valpredict))
```

### Making a pickle file 

```python
import joblib

# Save your trained model
joblib.dump(model, 'Parkinson_detector.pkl')
```
